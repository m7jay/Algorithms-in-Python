```python
"""
Queue implementation using lists
"""
class lQueue(object):
    def __init__(self):
        self.items = []
    
    def enque(self,value):
        self.items.insert(0, value)
    
    def deque(self):
        if self.items == []:
            raise Exception("Queue Empty!!!")
        return self.items.pop()
    
    def isEmpty(self):
        return self.items == []
    
    def size(self):
        return len(self.items)
    
    def peek(self):
        if self.items == []:
            raise Exception("Queue Empty!!!")
        return self.items[-1]

lq = lQueue()
print(lq.isEmpty())
lq.enque(1)
lq.enque(2)
lq.enque(3)
print(lq.size())
print(lq.peek())
print(lq.items)
print(lq.deque())
print(lq.deque())
print(lq.deque())
```

```python
"""
Queue implementation using two queues
"""
class sQueue(object):
    def __init__(self):
        self.in_stack = []
        self.out_stack = []
    
    def enque(self,value):
        self.in_stack.append(value)
    
    def deque(self):
        if self.out_stack:
            return self.out_stack.pop()
        while self.in_stack:
            self.out_stack.append(self.in_stack.pop())
        if not self.out_stack:
            raise Exception("Queue Empty!!!")
        return self.out_stack.pop()
    
    def size(self):
        return len(self.in_stack) + len(self.out_stack)

    def isEmpty(self):
        return (len(self.in_stack) + len(self.out_stack)) == 0
    
    def peek(self):
        if self.out_stack:
            return self.out_stack[-1]
        while self.in_stack:
            self.out_stack.append(self.in_stack.pop())
        
        if self.out_stack:
            return self.out_stack[-1]
        else:
            raise Exception("Queue Empty!!!")
            
sq = sQueue()
print(sq.isEmpty())
sq.enque(1)
sq.enque(2)
sq.enque(3)
print(sq.isEmpty())
print(sq.size())
print(sq.peek())
print(sq.deque())
print(sq.deque())
print(sq.deque())
#sq.deque()
```

```python
"""
Queue implementation using nodes
"""
class Node(object):
    def __init__(self, value):
        self.value = value
        self.next  = None
        
class nQueue(object):
    def __init__(self):
        self.front = None
        self.rear  = None
    def enque(self, value):
        new_node = Node(value)
        if self.front is None and self.rear is None:
            self.front = new_node
            self.rear  = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node
    
    def isEmpty(self):
        return self.front == None
    
    def deque(self):
        if self.isEmpty():
            raise Exception("Queue empty!!!")
        res = self.front.value
        self.front = self.front.next
        return res
    
    def size(self):
        temp = self.front
        count = 0
        while temp:
            count += 1
            temp = temp.next
        return count

    def peek(self):
        return self.front.value
    
nq = nQueue()
print(nq.size())
nq.enque(1)
nq.enque(2)
nq.enque(3)
print(nq.size())
print(nq.peek())
print(nq.deque())
print(nq.deque())
print(nq.deque())    
#print(nq.deque())
```
